```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('us-cities-top-1k-multi-year.csv')
print(df.head())
```

```python
states = ['New York', 'Connecticut', 'Pennsylvania', 'New Jersey']
filtered_df = df[df['State'].isin(states)]
grouped = filtered_df.groupby(['year', 'State'])['Population'].sum().reset_index()
sns.lineplot(data=grouped, x='year', y='Population', hue='State')
plt.title('Population by State (2014-2018)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.show()
```

```python
cities = ['Albany', 'Buffalo', 'Rochester', 'Yonkers', 'Syracuse']
filtered_df2 = df[(df['City'].isin(cities)) & (df['year'].isin([2014, 2018]))]
grouped = filtered_df2.groupby(['year', 'City'])['Population'].sum().reset_index()
sns.barplot(data=grouped, x='City', y='Population', hue='year')
plt.title('Population by City Comparing 2014 to 2018')
plt.ylabel('Total Population')
plt.xlabel('City')
plt.show()

```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

states = ['New York', 'Connecticut', 'Pennsylvania', 'New Jersey']
filtered_df = df[df['State'].isin(states)]
grouped = filtered_df.groupby(['year', 'State'])['Population'].sum().reset_index()
sns.lineplot(data=grouped, x='year', y='Population', hue='State', ax =ax1)
ax1.set_title('Population by State (2014-2018)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Population')

cities = ['Albany', 'Buffalo', 'Rochester', 'Yonkers', 'Syracuse']
filtered_df2 = df[(df['City'].isin(cities)) & (df['year'].isin([2014, 2018]))]
grouped = filtered_df2.groupby(['year', 'City'])['Population'].sum().reset_index()
sns.barplot(data=grouped, x='City', y='Population', hue='year')
ax1.set_title('Population by City Comparing 2014 to 2018')
ax2.set_ylabel('Total Population')
ax2.set_xlabel('City')

plt.tight_layout()
plt.show()
```
